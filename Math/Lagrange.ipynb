```python
import numpy as np
import plotly.graph_objects as go
```

```python
def join_xy(x, y, z_func):
    def map_x(_x):
        def map_y(y):
            return z_func(_x, y)
        return list(map(map_y, y))
    return list(map(map_x, x))
```

```python
def f_func(_x, _y):
    return _x**2*_y
def g_func(_x, _y):
    return _x**2+_y**2 - .3
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
g = np.array(join_xy(x, y, g_func))
f = np.array(join_xy(x, y, f_func))
```

```python
fig = go.Figure(data=[
        go.Surface(x = x, y = y, z = g, opacity=1, contours = { "z": {"show": True, "start": 0, "end": 2, "size": 1, "color":"blue"}}, colorscale='Electric'),
        go.Surface(x = x, y = y, z=f, opacity=1, contours = { "z": {"show": True, "start": 0, "end": 2, "size": 1, "color":"white"}}, colorscale='Viridis')
])
fig.update_layout(scene = { 'camera_eye': {"x": 0, "y": 0, "z": 1}, })
fig.show()
```
